```python
import random # random 함수를 구사하기 위해서 외부 패키지 불러오는 것
import copy   # 값 복사 함수 사용하려고 불러오는 것
```

```python
def initialization(x): #초기 세대 지정
    '''
    Create a chromosome with certain number of genes
    in this case, the created chromosome has 7 genes
    '''
    for j in range(0,7): # 7번 돌려서 저장
        x.append(random.randint(0,101))  
                  # 0부터 100 사이의 난수를 뽑아서 한번 씩 저장
    return x  # 이 로컬 함수 안에서만 통용되는 x
    # 다른 변수로 받을 수도 있기 때문에 return 함수 사용
```

```python
zcalchromosome=[]           # 함수값에 사용한 그 값
zcal=[]    # 적합도 함수값 계산 한 것 넣고

# zcalchromosome 리스트라고 지정, x값을 리스트로 받는데 식을 계산하기 위해서 원소 7개 가지고 있는 리스트 받아서 인덱싱
def fitnessfunction(x): #적합도 함수 식 설정 
    '''
    calculate the fitness function value of the chromosome
    (x1-10)**2+(x2-20)**2+(x3-50)**2+(x4-100)**2+(x5-70)**2'+(x6-40)**2+(x7-90)**2'''
    
    zcal.append((x[0]-10)**2+(x[1]-20)**2+(x[2]-50)**2+(x[3]-100)**2+(x[4]-70)**2+(x[5]-40)**2+(x[6]-90)**2)
    zcalchromosome.append(x)
    # x를 인풋으로 받아서 zcalchromosome에는 x를, 적합도 함수 값은 zcal에 추가
```

```python
cochromo=[]   # 크로스오버가 일어난 것을 저장
def crossover(a,b):#a,b 집합을 서로 셔플하는 함수를 설정
    """
    Two-point cross over 
    select the random point in list a and b  # 목록 a와 b에서 임의의 점을 선택
    change the elements between two points of a and b # a와 b의 두 점 사이의 원소 바꾸기

    a=[1,2,3,4,5]
    b=[6,7,8,9,10]
    
    >>> crossover(a,b)
    (if point=0,2)

    a=[6,7,8,4,5]
    b=[1,2,3,9,10]
    
      """
    twopointcount=[]    # 구간을 지정하는 투 포인트 인덱스 저장
    abcde=[]            # 이 리스트를 a에 넣고
    abcde=copy.copy(a)
    edcba=[]           #이 리스트를 b에 넣고
    edcba=copy.copy(b)
    
    
     #  랜덤추출 3번해서 twopointcount 리스트에 추가
    for i in range(3):
        twopointcount.append(random.randrange(0,len(a)))
              # 0부터 6까지 중에서 랜덤하게 하나를 뽑는다

              # ex. 111 - [1]   112  - [12]  123 - [123]
              # 3개를 지정한 이유 : 중복값을 없애기 위해서 

    twopointcount=list(set(twopointcount))
                # 구간 지정 리스트에 3번 넣어준다
    # 요소가 1개이면 2개까지 랜덤추출하여 요소 추가
    if len(twopointcount)==1:
        while len(twopointcount)<=2:
            twopointcount.append(random.randrange(0,len(a)))
            twopointcount=list(set(twopointcount))
    # 요소가 3개인 경우 3번째 요소 제거 
    if len(twopointcount)==3:
        del twopointcount[2]
        # 3개가 될 때까지 반복하면 while문이 break 하면서
        # 최종적으로 2개 값이 남게된다!!

    # twopointcount 섞은 후 정렬
    # a,b 수 edcba로 바꾸어줌
    # cochromo에 a,b 요소 추가
    random.shuffle(twopointcount)

# 그냥 바로 정렬해도 되지 않나,,,
    twopointcount.sort()
                                            #1, 2, 3, 4, 5   = a,    2, 3, 4, 4, 5
                                            # 2, 3, 4, 5, 6  = b     2, 3, 4, 5, 6
                                          # 이런 경우에 b는 바뀌지 않는다
                                          # 그래서 복사본을 다로 저장해서 crossover 에 사용한다

    a[twopointcount[0]:(twopointcount[1]+1)]=edcba[twopointcount[0]:(twopointcount[1]+1)]
    b[twopointcount[0]:(twopointcount[1]+1)]=abcde[twopointcount[0]:(twopointcount[1]+1)]
    cochromo.append(a)          # 맨처음 제시한 cochromo에 넣는다
    cochromo.append(b)
```

```python
def mutation(x): #돌연변이 함수 설정 
    mutindex=[] # mutation이 일어날 gene 후보를 담을 리스트 설정
    for i in range(0,len(x)):# 인풋으로 받은 crhomosome의 길이만큼 mutindex에 추가
        mutindex.append(i) #반복한 i값을 mutindex에 저장 
    random.shuffle(mutindex)#리스트 셔플
    x.pop(mutindex[0]) # 가장 앞의 index를 추출
    x.insert(mutindex[0],random.randint(0,101)) #첫번째 값(추출 후 비어 있는 자리)에 0부터 100 사이의 랜덤한 수를 넣음
    return x #생성한 돌연변이 값 반환

    # x = [1,2,3,4,5]
    # mutindex에 0, 1, 2, 3, 4 인덱스 값을 넣고 섞어서
    # ex. 4 3 2 1 0
    # mutindex [0] 에 들어온 인덱스를 찾아서 빼고
    # 0부터 100까지 난수를 뽑아서 넣어줍니다
```

```python
for i in range(0,4):
    z6=[]
    z6=(random.sample(z5,2)) # 룰렛에서 랜덤하게 2개 추출
    
    cmchromo=[]
    
    for i in range(0,len(z6)): # 랜덤으로 추출한 2개의 적합도 함숫값으로 chromosome을 찾아서 crossover와 mutation의 후보로 저장 
        for j in range(0,len(zt)): 
            if z6[i]==zt[j]:
                cmchromo.append(zcalchromosome[j])
                cmchromo=list(set(map(tuple,cmchromo)))    # 중복값을 제거하고 나면 원소가 tuple인 리스트임
                cmchromo=[list(elem) for elem in cmchromo] # 원소를 다시 리스트로 바꿔줌
              
                if len(cmchromo)<=1:    # 중복값이라 제거가 되면 똑같은 chromosome을 하나 더 넣어줌(비논리적)
                    cmchromo.append(cmchromo[0])

    corate=random.randint(0,100) #크로스오버확률이 될 난수를 불러옴
    murate=random.randint(0,100) #돌연변이확률이 될 난수를 불러옴 
    if corate>=30: # 70%의 확률로 crossover 수행
        crossover(cmchromo[0],cmchromo[1]) # 앞서 정의한 crossover함수로 crossover
        if murate>=5: # crossover를 수행했으면 95%의 확률로 mutation까지 수행
            for i in range(0,len(cmchromo)):  
                mutation(cmchromo[i])        # 두개의 크로모좀 다 mutation(1개의 gene값을 랜덤하게 변형) 발생시킴
                gdchromo.append(cmchromo[i]) # mutation한 chromosome을 저장
        else: # crossover가 수행되면 5%의 확률로 muation은 수행하지 않음
            for i in range(0,len(cmchromo)): 
                gdchromo.append(cmchromo[i]) # crossover만 수행한 chromosome을 저장

    else: # 30%확률로 crossover를 수행하지 않음(당연히 mutation도 수행하지 않음)
        for i in range(0,len(cmchromo)): # 후보였던 상태 그대로 chromosome을 저장
            gdchromo.append(cmchromo[i])

                                          # 이 과정을 총 여덟번 수행 (2 * 4)
                                          # len(gdchromo) == 10
```
