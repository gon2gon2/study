```python
import random
import copy
```

```python
# 리스트 x를 input으로 받아서 0~100 사이의 정수 7개를 x에 추가한 뒤 반환
def initialization(x):
    '''
    Create a chromosome with certain number of genes
    in this case, the created chromosome has 7 genes
    '''
    for j in range(0,7):
        x.append(random.randint(0,101))
    
    return x
```

```python
# x를 인풋으로 받아서 zcalchromosome에 추가한다
# x를 인풋으로 받아서 zcalchromosome에는 x를, 적합도 함수 값은 zcal에 추가

zcalchromosome=[]
zcal=[]
def fitnessfunction(x):
    '''
    calculate the fitness function value of the chromosome
    (x1-10)**2+(x2-20)**2+(x3-50)**2+(x4-100)**2+(x5-70)**2'+(x6-40)**2+(x7-90)**2'''
    
    zcal.append((x[0]-10)**2+(x[1]-20)**2+(x[2]-50)**2+(x[3]-100)**2+(x[4]-70)**2+(x[5]-40)**2+(x[6]-90)**2)
    
    zcalchromosome.append(x)
```

```python
cochromo=[]                 # crossover된 크로모좀
def crossover(a,b):         # chromosome a,b를 인풋으로 받는다
    """
    Two-point cross over
    select the random point in list a and b                # 리스트 a와 b를 인풋으로 받아서 각 리스트에서 하나의 포인트를 고른다
    change the elements between two points of a and b      # 그 두 점을 기준으로 a와 b의 원소를 바꿔줌

    a=[1,2,3,4,5]
    b=[6,7,8,9,10]
    
    >>> crossover(a,b)
    (if point=0,2)

    a=[6,7,8,4,5]
    b=[1,2,3,9,10]
    
    """
    twopointcount=[]        
    abcde=[]                 # 불필요한 코드
    abcde=copy.copy(a)       # 
    edcba=[]                 # 불필요한 코드
    edcba=copy.copy(b)       # 
    
    for i in range(3):    #  0,1,2,3,4 중에서 랜덤추출 3번해서 twopointcount 리스트에 추가
        twopointcount.append(random.randrange(0,len(a)))
    
    twopointcount=list(set(twopointcount))        # 중복값이 있을 수 있으니 set로 바꿨다가 list로 바꿔줌

    if len(twopointcount)==1:                   # 그 과정에서 3개가 똑같은 숫자일 경우에는 길이가 1이 돼서 2개가 될 때까지 반복
        while len(twopointcount)<=2:
            twopointcount.append(random.randrange(0,len(a)))
            twopointcount=list(set(twopointcount))
    
    if len(twopointcount)==3:                  # 3개가 다 다른 숫자가 나왔을 경우에는 길이가 3이라서 마지막 원소를 제거해줌
        del twopointcount[2]
        
    random.shuffle(twopointcount)              # twopointcount리스트를 랜덤하게 섞어줌

    twopointcount.sort()                       # twopoint를 정렬함  (교차지점1, 교차지점 2)
    # +1을 하는 이유: index가 연달아 있는 경우에는 crossover가 gene 하나에 대해서만 이뤄지기 때문에 +1을 해서 범위를 넓혀줌
    a[twopointcount[0]:(twopointcount[1]+1)]=edcba[twopointcount[0]:(twopointcount[1]+1)] # edcba(b의 카피)에서 바꾸려는 부분을 따와서 a를 바꿔줌
    b[twopointcount[0]:(twopointcount[1]+1)]=abcde[twopointcount[0]:(twopointcount[1]+1)] # abcde(a의 카피)에서 바꾸려는 부분을 따와서 b를 바꿔줌
    cochromo.append(a)  # 그렇게 바뀐 chromosome을 cochromo에 추가해줌
    cochromo.append(b)
```

```python
def mutation(x):
    mutindex=[]  # 빈 리스트 생성, mutation이 일어날 gene 후보를 담을 리스트
    for i in range(0,len(x)): # 인풋으로 받은 crhomosome의 길이만큼 mutindex에 추가
        mutindex.append(i)
    random.shuffle(mutindex)  # 랜덤하게 셔플
    x.pop(mutindex[0])        # 가장 앞의 index를 추출
    x.insert(mutindex[0],random.randint(0,101))  # 추출해서 비어있는 gene의 자리에 0~100 사이의 gene을 하나 넣어줌
    return x                 # shuffle하고 mutation이 일어난 chromosome x를 반환
```
