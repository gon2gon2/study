# 개선점


```python
target_dir='C:/projects/Data/animal'
```

```python
import os,shutil
targets=set()
fnames=os.listdir('C:/projects/Data/animal')
for fname in fnames:
    target = fname.split()[0]
    targets.add(target)
```

```python
for target in targets:
    # 새로운 경로 폴더 만들기
    new_dir = os.path.join(target_dir, target)
    try:
        os.mkdir(new_dir)
    except:
        pass
    
    # 파일 이름 찾기
    for fname in fnames:
        f_list = []
        if fname.startswith(target):
            f_list.append(fname)
        for f in f_list:
            src = os.path.join(target_dir, f)
            print(src)
            dst = os.path.join(new_dir, f)
            print(dst)
            shutil.copyfile(src, dst)
```
