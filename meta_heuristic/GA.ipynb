# Generate population

```python
# chromosome이 갖는 gene의 갯수와 population size를 입력 해주면 초기 population 리턴
def gen_pop(gene, pop_size):
    import numpy as np
    initial_pop = []
    for i in range(pop_size):
        np.random.randint(low=0, high=100, size=gene)
        initial_pop.append(list(np.random.randint(low=0, high=100, size=gene)))
    return initial_pop
```

```python
init_pop = gen_pop(7,10)
pd.DataFrame(init_pop)
```

# Fitness function

```python
# 적합도 함수, 하나의 chromosome에 대해 fitness value를 리턴
def fitness_function(lst):
    value = (lst[0]-10)**2 +    (lst[1]-20)**2 + (lst[2]-50)**2 +\
    (lst[3]-100)**2 +    (lst[4]-70)**2 +\
    (lst[5]-40)**2 +    (lst[6]-90)**2
    return value
```

```python
for i in init_pop:
    print(fitness_function(i))
```

# Selection
- 여기서는 룰렛휠 selection을 사용, 적합도에 비례해 선택확률을 할당
- 목표가 minimize니까 value가 작을수록 ratio 높음

## roullet wheel 함수
- population을 넣어주면 각 chromosome의 적합도 함수 값이 나오고, 그 값을 기반으로한 확률을 반환
- 이 확률을 저장해뒀다가 numpy, choice에 넣어주면 됨

```python
def roullet_wheel(pop):
    # 적합도 함수 계산
    values =[]
    for i in pop:
        values.append(fitness_function(i))
    # 각 녀석들의 비율 계산
    ratio = []
    hap = sum(values)
    for i in values:
        ratio.append(i/hap)

    return [values,ratio]
```

```python
val, ratio = roullet_wheel(init_pop)
print('각각의 적합도 함수 값:\n',val)
print()
print('그에 따른 확률 값:\n',ratio)
```

```python
def to_table(pop):
    values =[]
    for i in range(len(pop)):
        values.append(fitness_function(pop.iloc[i,:]))
        
        
    table = pd.DataFrame(pop)
    table['fitness_value'] = values
    table['rank'] = table['fitness_value'].rank(method='min', ascending=False)
    

    hap = sum(table['fitness_value'])
    
    table['ratio'] = table['fitness_value']/sum(table['fitness_value'])
    return table
```
